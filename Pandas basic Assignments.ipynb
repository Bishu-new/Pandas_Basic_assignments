Ans 4 : 

**Series**: A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index. It is similar to a column in a spreadsheet or a vector in mathematical terms. Here's an example of creating a Series:

```python
import pandas as pd
s = pd.Series([1, 3, 5, 6, 8])
print(s)
    
```

**DataFrame**: A DataFrame is a two-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dictionary of Series objects. It is generally the most commonly used pandas object. Here's an example of creating a DataFrame:

```python
df = pd.DataFrame({'A': 1.,
                       'B': pd.Timestamp('20130102'),
                       'C': pd.Series(1, index=list(range(4)), dtype='float32'),
                       'D': pd.array([3] * 4, dtype='int32'),
                       'E': pd.Categorical(["test", "train", "test", "train"]),
                       'F': 'foo'})
print(df)
```

Ans 5 : Pandas provides a variety of functions to manipulate data in a DataFrame. Here are some common ones:

- **head()**: This function is used to get the first n rows. This is useful for quickly testing if your object has the right type of data in it. For example, `df.head(5)` will return the first 5 rows of the DataFrame df.

- **describe()**: This function generates descriptive statistics that summarize the central tendency, dispersion, and shape of a dataset’s distribution, excluding NaN values. For example, `df.describe()` will provide count, mean, standard deviation, minimum and maximum values along with 25th, 50th, and 75th percentiles.

- **loc[]**: This function is used to access a group of rows and columns by label(s) or a boolean array. For example, `df.loc[:, 'A']` will return all rows for column 'A'.

- **iloc[]**: This function is purely integer-location based indexing for selection by position. For example, `df.iloc[0]` will return the first row.

- **drop()**: This function is used to drop specified labels from rows or columns. For example, `df.drop('A', axis=1)` will drop column 'A' from df.

- **groupby()**: This function is used to split the data into groups based on some criteria. For example, `df.groupby('A').sum()` will return the sum of all other columns for each unique value in column 'A'.

Ans 6 : All three pandas data structures - Series, DataFrame, and Panel - are mutable in nature. This means that they can be modified after they are created. For example, you can change the values of a Series, add or remove columns from a DataFrame, or update the items of a Panel. Here's an example of modifying a DataFrame:

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# Modify the DataFrame
df.loc[0, 'A'] = 100

# Print the modified DataFrame
print("Modified DataFrame:")
print(df)
```

Ans 7 : Sure, we can create a DataFrame using multiple Series in pandas. Here’s an example:

```python
import pandas as pd

# Create a Series
s1 = pd.Series([1, 2, 3, 4], name='s1')
s2 = pd.Series(['a', 'b', 'c', 'd'], name='s2')

# Create a DataFrame using the Series
df = pd.concat([s1, s2], axis=1)

# Print the DataFrame
print(df)
```
